# Importing Libraries + Setting Path + Importing data frame

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy
path = r'C:\Users\Chaot\Python stuff\06 - Utah Real Estate Housing Prices'
```

```python
clean_cities = pd.read_csv(os.path.join(path, 'Data Brief + Data Sets', 'Utah Housing Datasets', 'cleaned_cities.csv'), index_col = False)
clean_cities
```

```python
clean_cities.dtypes
```

```python
clean_cities['price'] = clean_cities['price'].astype(int)
clean_cities['bedrooms'] = clean_cities['bedrooms'].fillna(0).astype(int)
clean_cities['bathrooms'] = clean_cities['bathrooms'].fillna(0).astype(int)
clean_cities['zestimate'] = clean_cities['zestimate'].fillna(0).astype(int)
clean_cities['tax_assessed_value'] = clean_cities['tax_assessed_value'].fillna(0).astype(int)
clean_cities['new_home'] = clean_cities['new_home'].fillna(False).astype(bool)
```

```python
clean_cities
```

```python
cities_relationship = clean_cities[['zipcode','price','bathrooms',
                                   'bedrooms','living_area']]
```

```python
# Revisting these graphs to exclude the high outliers which is above 2,037,500
cities_relationship = cities_relationship[cities_relationship['price'] < 2037500]
```

```python
cities_relationship.corr()
```

```python
f,ax = plt.subplots(figsize=(9,9))

cc_corr = sns.heatmap(cities_relationship.corr(), annot = True, ax = ax)
plt.title('Heat map and Correlation Matrix', fontsize=14)
plt.savefig(os.path.join(path, 'Visualizations', 'Heat map and Correlation Matrix.png'),bbox_inches='tight')
```

From the above heatmap and correlation matrix  we see that zipcode has a weak correlation strength to everything. Our most important factor that we're interested in, price, has a strong relationship with bathrooms and a weak relationship with bedrooms. It does, however, also have a strong relationship with living area. 

# Scatterplot Relationships.

```python
la_p = sns.lmplot(x = 'living_area', y = 'price', data = cities_relationship, line_kws={'color': 'red'})
plt.title('Price to Living Area', fontsize=14)
plt.ylabel('Price (in Millions)')
plt.xlabel('Living Area')
la_p.figure.savefig(os.path.join(path, 'Visualizations', 'la_pr_scatter.png'),bbox_inches='tight')
```

Here we see that most housesincrease in price as living area increases. There are a few outliers however showing abnormally high prices at certain living areas.

```python
sns.histplot(cities_pair['bathrooms'], bins = 15, kde = True)
plt.title('Bathrooms Histogram')
plt.show()
```

Checked three options to pllot against a price plot. Going with bedrooms since it seems logical for this categorization..

```python
cities_pair.loc[cities_pair['bedrooms'] <=2, 'house_size'] = 'Small house'
```

```python
cities_pair.loc[cities_pair['bedrooms'] >=5, 'house_size'] = 'Large house'
```

```python
cities_pair.loc[(cities_pair['bedrooms'] > 2) & (cities_pair['bedrooms'] < 5), 'house_size'] = 'Average house'
```

```python
cities_pair['house_size'].value_counts(dropna = False)
```

```python
# Creating categorical plot
sns.set(style="ticks")
cat_price_bath = sns.catplot(x='price', y='bathrooms', hue="house_size", data=cities_pair)
plt.title('Price & Bathroom Scatterplot categorized by House Size')
cat_price_bath.figure.savefig(os.path.join(path, 'Visualizations', 'Price & Bathroom Categorized Scatter x House Size.png'),bbox_inches='tight')
```

```python
# Creating categorical plot
sns.set(style="ticks")
cat_area_bath = sns.catplot(x='price', y='bedrooms', hue="house_size", data=cities_pair)
plt.title('Price & Bedroom Scatterplot categorized by House Size')
cat_area_bath.figure.savefig(os.path.join(path, 'Visualizations', 'Price & Bedroom Categorized Scatter x House Size.png'),bbox_inches='tight')
```

```python
# Creating categorical plot
sns.set(style="ticks")
cat_bed_bath = sns.catplot(x='price', y='living_area', hue="house_size", data=cities_pair)
plt.title('Price & Living Area Scatterplot categorized by House Size')
cat_bed_bath.figure.savefig(os.path.join(path, 'Visualizations', 'Price & Living Area Categorized Scatter x House Size.png'),bbox_inches='tight')
```

As seen in the above categorical plots, prices are increasing based on the house size. What is surprising here is that each size of house actually varies along our price axis. Even small and average houses range from the lower end price to higher end price. This could possibly mean that the smaller and average houses that are more expensive are newer. It might also be possible that the home type is playing a part (Single Family, Condo, Townhouse).
