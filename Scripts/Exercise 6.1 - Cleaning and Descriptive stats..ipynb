# Importing Libraries + Setting Path + Importing dataset.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
path = r'C:\Users\Chaot\Python stuff\06 - Utah Real Estate Housing Prices'
```

```python
og_cities = pd.read_csv(os.path.join(path, 'Data Brief + Data Sets', 'Utah Housing Datasets', 'Consolidated Cities.csv'), index_col = False)
pd.set_option('display.max_columns',100)
og_cities.head(10)
```

```python
og_cities.info()
```

```python
# Dropping these columns. They don't seem relevant to what we need.
og_cities = og_cities.drop(columns = ['homeStatus','isPreforeclosureAuction','daysOnZillow','isZillowOwned',
                                      'lotAreaValue','lotAreaUnit','is_openHouse','openHouse','open_house_showing',
                                      'imgSrc','currency','price_to_rent_ratio','rentZestimate','datePriceChanged', 'streetName',
                                      'priceReduction','priceChange'])
og_cities.info()
```

```python
og_cities.rename(columns = {'livingArea':'living_area',
                           'homeType':'home_type',
                           'taxAssessedValue':'tax_assessed_value',
                           'is_newHome':'new_home',
                           'newConstructionType':'construction_type',
                           'homeDetailUrl':'home_detail_link'}, inplace = True)
og_cities.head(1)
```

```python
# Noticing that listings have MULTI_FAMILY as a home type aren't very good, removing these rows from the data set.

og_cities = og_cities.drop(og_cities[og_cities['home_type'] == 'MULTI_FAMILY'].index)

og_cities['home_type'].value_counts()
```

```python
og_cities.info()
```

```python
# Checking City values to see if they all match, no capitals or abbreviations.

og_cities['city'].value_counts()
```

```python
# checking to see if there are duplicate listings.

og_cities_dupe = og_cities[og_cities.duplicated()]
og_cities_dupe
```
